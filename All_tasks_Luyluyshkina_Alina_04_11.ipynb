```python
def binary_search_sqrt(target):
    l = 0
    r = target

    while l <= r:
        middle = l + (r - l) // 2
        sq = middle * middle

        if sq > target:
            r = middle - 1
        elif sq < target:
            l = middle + 1
        else:
            return middle

    return r
print(binary_search_sqrt(25))
```

```python
def copy_time(n, x, y):
    l = 0
    r = (n - 1) * max(x, y)

    while l + 1 < r:
        mid = (l + r) // 2
        if mid // x + mid // y < n - 1:
            l = mid
        else:
            r = mid

    return r + min(x, y)

print(copy_time(5, 1, 2))
```

```python
import math
def copy_time_math(n, x, y):
    if n == 0:
        return 0
    if n == 1:
        return min(x, y)

    fast = min(x, y)
    time = math.ceil((n - 1) * x * y / (x + y)) # составила уравнение t/x + t/y = n-1 и выразила из него t = (n-1)*x * y / (x + y)

    return fast + time

print(copy_time_math(5, 1, 2))
```

```python
def feed_animals(animals, food):
    if len(animals) == 0 or len(food) == 0:
        return 0

    animals.sort()
    food.sort()

    count = 0
    for f in food:
        if f >= animals[count]:
            count += 1
        if count == len(animals):
            break

    return count

animals = [8, 2, 3, 2]
food = [1, 4, 3, 8]
print(feed_animals(animals, food))
```

```python
def extra_letter(a, b):
    hash_map_a = {}

    for char in a:
        hash_map_a[char] = hash_map_a.get(char, 0) + 1

    for char in b:
        if char in hash_map_a:
            hash_map_a[char] -= 1
            if hash_map_a[char] == 0:
                del hash_map_a[char]
            continue

        return char

    return ""

print(extra_letter("abcd", "abcde"))
print(extra_letter("abc", "abc"))
```

```python
def two_sum(data, target):
    cache = {}

    for i in range(len(data)):
        cache[data[i]] = i

    for i in range(len(data)):
        diff = target - data[i]
        if diff in cache and i != cache[diff]:
            return [i, cache[diff]]

    return []

print(two_sum([2, 10, 11, 15], 13))
print(two_sum([1, 2, 3], 10))
```

```python
import timeit
def shell_sort(arr):
    n = len(arr)
    gap = n // 2

    while gap > 0:
        for current_position in range(gap, n):
            m_gap = current_position
            while m_gap >= gap and arr[m_gap] < arr[m_gap - gap]:

                arr[m_gap], arr[m_gap - gap] = arr[m_gap - gap], arr[m_gap]
                m_gap -= gap
        gap = gap // 2

    return arr

arr1 = [64, 34, 25, 12, 22, 11, 90]
print(shell_sort(arr1))

def benchmark_shell_sort():
    test_arr = arr1.copy()
    return shell_sort(test_arr)


time = timeit.timeit(benchmark_shell_sort, number=1000)
print(f"Shell Sort время: {time:.6f}с ")
```

```python
def insert_sort(arr):

    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key

    return arr

arr1 = [64, 34, 25, 12, 22, 11, 90]
print(shell_sort(arr1))

def benchmark_insert_sort():
    test_arr = arr1.copy()
    return insert_sort(test_arr)


time = timeit.timeit(benchmark_insert_sort, number=1000)
print(f"Insert sort время: {time:.6f}с ") #сортировка вставками быстрее?
```

```python
def group_anagrams(strs):
    groups = {}

    for word in strs:
        sorted_word = ''.join(sorted(word))

        if sorted_word not in groups:
            groups[sorted_word] = []
        groups[sorted_word].append(word)

    return list(groups.values())

strok = ["eat","tea","tan","ate","nat","bat"]
print(group_anagrams(strok))
```
