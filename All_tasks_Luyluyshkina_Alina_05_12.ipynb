Task_1 (Последовательность из 0 и 1)

```python
def b_sequences(n):
  if n == 1:
    return 2
  if n == 2:
    return 3

  return b_sequences(n - 1) + b_sequences(n - 2)

print(b_sequences(3))
```

```python
def b_sequences_dynamic(n):
    if n == 0:
        return 1
    if n == 1:
        return 2

    dp = [1, 2]
    i = 2
    while i <= n:
        dp.append(dp[i - 1] + dp[i - 2])
        i += 1

    return dp[n]

print(b_sequences_dynamic(3))
```

Task_2 (Последовательность из 3-х единиц)

```python
def b_sequences_dynamic_3(n):
    if n == 0:
        return 1
    if n == 1:
        return 2
    if n == 2:
        return 4

    dp = [1, 2, 4]
    i = 3
    while i <= n:
        dp.append(dp[i - 1] + dp[i - 2] + dp[i - 3])
        i += 1

    return dp[n]

print(b_sequences_dynamic_3(4))
```

Task_3 (Наибольшая непрерывная возрастающая последовательность)

```python
def findLIS(nums):
    if len(nums) == 0:
        return 0

    dp = [1] * len(nums)

    for i in range(len(nums)):
        for j in range(i):
            if nums[j] < nums[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1

    return max(dp)

nums_1 = [3, 1, 4, 7, 2, 11, 9]
print(findLIS(nums_1))
```

Task_4 (Треугольник Паскаля)

```python
def pascal_triangle(row, col):

    if col == 0 or row == col:
        return 1
    else:
        return pascal_triangle(row - 1, col - 1) + pascal_triangle(row - 1, col)

n = 6
dp = []
for row in range(n):
    currentRow = []
    for col in range(row + 1):
        currentRow.append(pascal_triangle(row, col))
    dp.append(currentRow)

print("\nТреугольник Паскаля:")
for row in dp:
    print(row)
```

```python
def pascal_triangle_iterative(n):
    dp = []
    for i in range(n):
        tmp = []
        for j in range(i + 1):
            tmp.append(1)
        dp.append(tmp)

    for row in range(2, n):
        for col in range(1, row):
            dp[row][col] = dp[row - 1][col - 1] + dp[row - 1][col]

    return dp

n = 6
dp = pascal_triangle_iterative(n)

print("\nТреугольник Паскаля:")
for row in dp:
    print(row)
```

Task_5 (Максимальная выгода)

```python
def maxProfit(prices):
    if len(prices) < 2:
        return 0

    profit = 0
    min_price = prices[0]

    for current_price in prices[1:]:
        profit = max(profit, current_price - min_price)
        min_price = min(current_price, min_price)

    return profit

prices = [8, 9, 3, 7, 4, 16, 12]
print(maxProfit(prices))
```

Task_6 (Размен монет)

```python
def coinChange(coins, amount, cache=None):
    if cache is None:
        cache = {}

    if amount == 0:
        return 0
    if amount < 0:
        return -1

    if amount in cache:
        return cache[amount]

    minCoins = amount + 1

    for coin in coins:
        res = coinChange(coins, amount - coin, cache)
        if res >= 0 and res + 1 < minCoins:
            minCoins = res + 1

    if minCoins == amount + 1:
        cache[amount] = -1
    else:
        cache[amount] = minCoins

    return cache[amount]

coins = [1, 2, 5]
amount = 11
print(f"Минимальное количество монет для суммы {amount}: {coinChange(coins, amount)}")
```

```python
def coinChange(coins, amount):
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0

    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], dp[i - coin] + 1)

    if dp[amount] == float('inf'):
        return -1
    else:
        return dp[amount]

coins = [1, 2, 5]
amount = 11
print(f"Минимальное количество монет для суммы {amount}: {coinChange(coins, amount)}")
```

Task_7 (Максимальный палиндром в строке)

```python
def longestPalindrome(s):
    if not s:
        return ""

    currentMaxLeft = 0
    currentMaxRight = 0

    def expandAroundCenter(l, r):
        nonlocal currentMaxLeft, currentMaxRight
        while l >= 0 and r < len(s) and s[l] == s[r]:
            if (r - l) > (currentMaxRight - currentMaxLeft):
                currentMaxRight = r
                currentMaxLeft = l
            l -= 1
            r += 1

    for i in range(len(s)):
        expandAroundCenter(i, i)
        expandAroundCenter(i, i + 1)

    return s[currentMaxLeft:currentMaxRight + 1]

test_case = "babad"
result = longestPalindrome(test_case)
print(f"Строка: '{test_case}' самый длинный палиндром: '{result}'")
```

```python
def longestPalindrome(s):
    if not s:
        return ""

    n = len(s)
    dp = [[False] * n for _ in range(n)]

    for i in range(n):
        dp[i][i] = True

    start = 0
    max_len = 1

    for i in range(n - 1):
        if s[i] == s[i + 1]:
            dp[i][i + 1] = True
            start = i
            max_len = 2

    for length in range(3, n + 1):
        for i in range(n - length + 1):
            j = i + length - 1

            if s[i] == s[j] and dp[i + 1][j - 1]:
                dp[i][j] = True

                if length > max_len:
                    start = i
                    max_len = length

    return s[start:start + max_len]

test_case = "babad"
result = longestPalindrome(test_case)
print(f"Строка: '{test_case}' самый длинный палиндром: '{result}'")
```
