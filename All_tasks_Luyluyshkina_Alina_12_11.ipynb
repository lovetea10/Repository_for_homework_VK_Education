Task_1 (Восстановление бинарного дерева из массива)

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right

def build_tree(arr, i):
    if i >= len(arr) or arr[i] is None:
        return None

    root = TreeNode(arr[i])
    root.left = build_tree(arr, 2 * i + 1)
    root.right = build_tree(arr, 2 * i + 2)

    return root
```

Task_2 (Симметричное дерево)

```python
def is_symmetric(root):
    if root is None:
        return True

    queue = [root]

    while queue:
        queue_len = len(queue)
        for i in range(queue_len):
            left_node = queue[i]
            right_node = queue[queue_len - i - 1]

            if left_node is None and right_node is None:
                continue
            if left_node is None or right_node is None:
                return False
            if left_node.data != right_node.data:
                return False

        next_level = []
        for node in queue:
            if node:
                next_level.append(node.left)
                next_level.append(node.right)
            else:
                next_level.append(None)
                next_level.append(None)

        if all(node is None for node in next_level):
            break

        queue = next_level

    return True
root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

print(is_symmetric(root1))
```

```python
def dept_search(root, res):
    if root is None:
        return res

    dept_search(root.left, res)
    res.append(root.data)
    dept_search(root.right, res)

    return res

def is_symmetric_dfs(root):
    if root is None:
        return True

    data = []
    data = dept_search(root, data)

    left, right = 0, len(data) - 1
    while left < right:
        if data[left] != data[right]:
            return False
        left += 1
        right -= 1

    return True

root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

print(is_symmetric_dfs(root1))
```

Task_3 (Поиск минимальной глубины бинарного дерева)

```python
def min_depth_dfs(root):

    if root is None:
        return 0

    if root.left is None and root.right is None:
        return 1

    if root.left is None:
        return min_depth_dfs(root.right) + 1

    if root.right is None:
        return min_depth_dfs(root.left) + 1

    return min(min_depth_dfs(root.left), min_depth_dfs(root.right)) + 1

root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

print(min_depth_dfs(root1))
```

Task_4 (Поиск максимального произведения, по идее ищем то 2 макс значения и их перемножаем? Но на семинаре ищем мин и макс занчения, поэтому реализация с семинара)

```python
def max_min_multiplication(data):

    if len(data) < 3:
        return -1

    min_index = 1
    max_index = 2
    i = 0

    while True:
        min_index_tmp = 2 * i + 1
        if min_index_tmp < len(data):
            min_index = min_index_tmp
            i = min_index_tmp
            continue
        break

    result = data[min_index] * data[max_index]
    return result
```

Task_5 (Являются ли два дерева одинаковыми)

```python
def is_same_tree(a, b):

    if a == None and b == None:
        return True
    if a == None or b == None:
        return False
    if a.data != b.data:
        return False

    return is_same_tree(a.left, b.left) and is_same_tree(a.right, b.right)

root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

root2 = TreeNode(1)
root2.left = TreeNode(2, TreeNode(3), TreeNode(4))
root2.right = TreeNode(2, TreeNode(4), TreeNode(3))

print(is_same_tree(root1, root2))
```
