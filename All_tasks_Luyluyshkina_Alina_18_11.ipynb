Task_1 (Проверка  корректности кучи)

```python
def is_max_heap(arr):
    n = len(arr)
    for i in range(n):
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n and arr[i] < arr[left]:
            return False
        if right < n and arr[i] < arr[right]:
            return False

    return True

arr1 = [100, 90, 80, 50, 85, 70]
arr2 = [10, 20, 30]

print(is_max_heap(arr1))
print(is_max_heap(arr2))
```

```python
from collections import deque

def is_max_heap_bfs(arr):
    n = len(arr)
    if n <= 1:
        return True

    queue = deque()
    queue.append(0)

    while len(queue) > 0:
        i = queue.popleft()
        left = 2 * i + 1
        right = 2 * i + 2

        if left < n:
            if arr[i] < arr[left]:
                return False
            queue.append(left)

        if right < n:
            if arr[i] < arr[right]:
                return False
            queue.append(right)

    return True

arr1 = [100, 90, 80, 50, 85, 70]
arr2 = [10, 20, 30]

print(is_max_heap_bfs(arr1))
print(is_max_heap_bfs(arr2))
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.data = val
        self.left = left
        self.right = right
        self.balance_factor = 0
```

```python
from collections import deque

def is_max_heap_root(root):
    if not root:
        return True

    queue = deque([root])
    should_be_leaf = False

    while queue:
        current = queue.popleft()

        if current.left:
            if should_be_leaf or current.left.data > current.data:
                return False
            queue.append(current.left)
        else:
            should_be_leaf = True

        if current.right:
            if should_be_leaf or current.right.data > current.data:
                return False
            queue.append(current.right)
        else:
            should_be_leaf = True

    return True


root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

root2 = TreeNode(100,TreeNode(90, TreeNode(50), TreeNode(85)),
            TreeNode(80, TreeNode(70)))

print(is_max_heap_root(root1))
print(is_max_heap_root(root2))
```

Task_2 (Проверка на полное бинарное дерево)

```python
from collections import deque

def is_сomplete_tree(root):
    if not root:
        return True

    queue = deque([root])
    should_be_leaf = False

    while queue:
        node = queue.popleft()

        if not node:
            should_be_leaf = True
        else:
            if should_be_leaf:
                return False
            queue.append(node.left)
            queue.append(node.right)

    return True

root1 = TreeNode(1)
root1.left = TreeNode(2, TreeNode(3), TreeNode(4))
root1.right = TreeNode(2, TreeNode(4), TreeNode(3))

root2 = TreeNode(100,TreeNode(90, TreeNode(50), TreeNode(85)),
            TreeNode(80, TreeNode(70)))

root3 = TreeNode(100,
                TreeNode(90, TreeNode(50), TreeNode(85)),
                TreeNode(80, None, TreeNode(70)))

print(is_сomplete_tree(root1))
print(is_сomplete_tree(root2))
print(is_сomplete_tree(root3))
```

Task_3 (Объединение К отсортированных массивов)

```python
import heapq

def merge_K_sorted_arrays(sorted_arrays):
    min_heap = []

    for arr in sorted_arrays:
        for element in arr:
            heapq.heappush(min_heap, element)

    merged_array = []
    while min_heap:
        merged_array.append(heapq.heappop(min_heap))

    return merged_array

arrays1 = [
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9]
]
result1 = merge_K_sorted_arrays(arrays1)
print(result1)
```

```python
import heapq #надо для работы с кучей в питоне

def merge_K_sorted_arrays_eff(sorted_arrays):

    merged_array = []
    min_heap = []

    for i, current_array in enumerate(sorted_arrays):
        if len(current_array) > 0:
            heapq.heappush(min_heap, (current_array[0], i, 0))

    while min_heap:
        value, array_idx, element_idx = heapq.heappop(min_heap)
        merged_array.append(value)

        if element_idx + 1 < len(sorted_arrays[array_idx]):

            next_element = sorted_arrays[array_idx][element_idx + 1]
            heapq.heappush(min_heap, (next_element, array_idx, element_idx + 1))

    return merged_array

arrays1 = [
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9]
]
result1 = merge_K_sorted_arrays_eff(arrays1)
print(result1)
```

Task_4 (К-ый наименьший/наибольший элемент)

```python
def inorder_min_iterative(node, k):
    stack = []
    counter = 0
    current = node

    while len(stack) > 0 or current is not None:

        while current is not None:
            stack.append(current)
            current = current.left

        current = stack.pop()
        counter += 1

        if counter == k:
            return current.data

        current = current.right

    return None

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)
root.left.left.left = TreeNode(0)

for k in range(1, 8):
    result = inorder_min_iterative(root, k)
    print(f"{k}-й наименьший элемент: {result}")
```

```python
def inorder_min(node, k, counter):
    if node is None:
        return None

    left_result = inorder_min(node.left, k, counter)

    if left_result is not None:
        return left_result

    counter[0] += 1

    if counter[0] == k:
        return node.data

    return inorder_min(node.right, k, counter)

root = TreeNode(5)
root.left = TreeNode(3)
root.right = TreeNode(7)
root.left.left = TreeNode(2)
root.left.right = TreeNode(4)
root.right.right = TreeNode(8)
root.left.left.left = TreeNode(1)

def find_kth_smallest(root, k):
    counter = [0]
    return inorder_min(root, k, counter)

for k in range(1, 9):
    result = find_kth_smallest(root, k)
    print(f"{k}-й наименьший элемент: {result}")
```

Task_5 (Balance factor)

```python
def calculate_heights_and_balance(node):
    if node is None:
        return 0

    left_height = calculate_heights_and_balance(node.left)
    right_height = calculate_heights_and_balance(node.right)

    node.balance_factor = left_height - right_height

    return 1 + max(left_height, right_height)

root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.right = TreeNode(6)
root.left.left.left = TreeNode(7)

height = calculate_heights_and_balance(root)

def print_tree_with_balance(node, level=0):
    if node:
        print("  " * level + f"Node {node.data} (balance: {node.balance_factor})")
        print_tree_with_balance(node.left, level + 1)
        print_tree_with_balance(node.right, level + 1)

print_tree_with_balance(root)
print(f"\nОбщая высота дерева: {height}")
```

Task_6 (Преобразование в зеркальное дерево)

```python
def mirror_tree_iterative(node):
    if node is None:
        return None

    queue = deque([node])

    while queue:
        current = queue.popleft()

        temp = current.left
        current.left = current.right
        current.right = temp

        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)

    return node
```

```python
def mirror_tree_recursive(node):
    if node is None:
        return None

    left_mirrored = mirror_tree_recursive(node.left)
    right_mirrored = mirror_tree_recursive(node.right)

    node.left = right_mirrored
    node.right = left_mirrored

    return node
```
